## PAGOS A PROVEEDORES Y CONTRATISTAS, GOBIERNO DE MONTERREY 2018

```python
%matplotlib inline
import pandas as pd
```

```python
compras_df = pd.read_csv('clean_compras_2018.csv')
```

```python
compras_df.head(5)
```

```python
data=compras_df.copy()
```

## Llenar los valores faltantes del dataframe

```python
#utiliza el for con un indice para en una lista almacenar el tipo de dato en cadena
for i, column in enumerate(list([str(d) for d in data.dtypes])):
    #si el tipo de dato es object
    if column == "object":
        data[data.columns[i]] = data[data.columns[i]].fillna(data[data.columns[i]].mode())
        data[data.columns[i]] = data[data.columns[i]].astype("category").cat.codes
    else:
        data[data.columns[i]] = data[data.columns[i]].fillna(data[data.columns[i]].median())
```

```python
data.head(5)
```

## Utilizar random forest

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.model_selection import train_test_split
```

```python
y = data.pop('IMPORTE')
```

```python
y
```

```python
x = data.copy()
```

```python
x 
```

## Dividir los datos en dos grupos

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
clf = RandomForestRegressor(n_estimators=1000, max_depth = 4)
```

```python
clf.fit(x_train,y_train)
```

## Coeficiente de Determinación

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, clf.predict(x_test))
```

## modelo no supervisado

```python
import warnings
warnings.filterwarnings('ignore', category=UserWarning, module='sklearn')
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
compras_df_c=compras_df.copy()
```

```python
compras_df_c.head(5)
```

```python
# Agrupar por beneficiario con las funciones especificadas
compras_df_c_grupo = compras_df_c.groupby('BENEFICIARIO').agg({
    'CONTRATO': lambda x: x.nunique(), 
    'IMPORTE': 'sum',
    'FACTURA': lambda x: x.nunique()
}).reset_index()
```

```python
compras_df_cluster=compras_df_c_grupo
```

```python
compras_df_cluster
```

```python
compras_df_c_grupo_f=compras_df_c_grupo.drop(['BENEFICIARIO'], axis=1)
```

```python
compras_df_c_grupo_f
```

```python
normalizado_compras_df=(compras_df_c_grupo_f-compras_df_c_grupo_f.min())/(compras_df_c_grupo_f.max()-compras_df_c_grupo_f.min())
```

```python
normalizado_compras_df
```

```python
wcss=[]
```

```python
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, max_iter=300, random_state=42)
    kmeans.fit(normalizado_compras_df)
    wcss.append(kmeans.inertia_)
```

```python
# Graficar el método del codo
plt.figure(figsize=(8,6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--', color='b')
plt.title('Codo de Jambu: Selección del número óptimo de clústeres')
plt.xlabel('Número de clústeres')
plt.ylabel('WCSS (Within-Cluster Sum of Squares)')
plt.grid(True)
plt.show()
```

```python
clustering = KMeans(n_clusters=3, max_iter=300, random_state=42)
```

```python
clustering.fit(normalizado_compras_df)
```

```python
compras_df_cluster['KMeans_Cluster']=clustering.labels_
```

```python
compras_df_cluster['KMeans_Cluster'].value_counts()
```

## ANALISIS DE COMPONENTES PRINCIPALES

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 2)
```

```python
pca_compras = pca.fit_transform(normalizado_compras_df)
```

```python
pca_compras_df = pd.DataFrame(data=pca_compras,columns = ['componentes_1','componentes_2'])
```

```python
pca_nombres_beneficia = pd.concat([pca_compras_df,compras_df_cluster['KMeans_Cluster']],axis = 1)
```

```python
pca_nombres_beneficia
```

```python
# Graficar los clústeres
plt.figure(figsize=(10,8))

scatter = plt.scatter(pca_nombres_beneficia['componentes_1'],
                      pca_nombres_beneficia['componentes_2'],
                      c=pca_nombres_beneficia['KMeans_Cluster'],
                      cmap='viridis',
                      s=100,
                      alpha=0.7)

# Añadir etiquetas y título
plt.title('Gráfico de Clústeres KMeans con PCA')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.colorbar(scatter)
plt.grid(True)
plt.show()
```
