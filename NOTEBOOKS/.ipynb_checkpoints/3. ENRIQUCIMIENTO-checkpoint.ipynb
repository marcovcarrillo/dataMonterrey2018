## PAGOS A PROVEEDORES Y CONTRATISTAS, GOBIERNO DE MONTERREY 2018

```python
%matplotlib inline
import pandas as pd
```

```python
compras_df = pd.read_csv('clean_compras_2018.csv')
```

```python
ls
```

```python
compras_df.head(4)
```

```python
compras_df.dtypes
```

```python
compras_df['FECHA']=pd.to_datetime(compras_df['FECHA'])
```

```python
compras_df['DIA']=compras_df['FECHA'].apply(lambda x: x.day)
```

```python
compras_df['MES']=compras_df['FECHA'].apply(lambda x: x.month)
```

```python
compras_df['ANIO']=compras_df['FECHA'].apply(lambda x: x.year)
```

```python
compras_df.dtypes
```

```python
compras_df.head(5)
```

## Proyectar una variable con otra

```python
beneficiario_sales=compras_df['BENEFICIARIO'].value_counts().to_dict()
```

```python
list(beneficiario_sales.keys())[0:5]
```

```python
compras_df['BENEFICIARIO_SALES']=compras_df['BENEFICIARIO'].apply(lambda x: beneficiario_sales[x])
```

```python
compras_df.head(4)
```

```python
sheet_mean_compras=compras_df.groupby('ETIQUETA')['IMPORTE'].mean()
```

```python
sheet_mean_compras
```

```python
sheet_mean_compras.to_dict()
```

```python
compras_df['IMPORTE_GASTO_PROMEDIO']=compras_df['ETIQUETA'].apply(lambda x: sheet_mean_compras[x])
```

```python
compras_df.to_csv(r'C:\Users\WINDOWS\Documents\INTELIGENCIA ARTIFICIAL\PROYECTO\compras_df_enriched.csv', index=None)
```
