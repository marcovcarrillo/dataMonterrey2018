```python
compras_df['FECHA']=pd.to_datetime(compras_df['FECHA'])
```

```python
compras_df['DIA']=compras_df['FECHA'].apply(lambda x: x.day)
```

```python
compras_df['MES']=compras_df['FECHA'].apply(lambda x: x.month)
```

```python
compras_df['ANIO']=compras_df['FECHA'].apply(lambda x: x.year)
```

```python
compras_df= compras_df[['PROVINCIA', 'BENEFICIARIO', 'RFC', 'FACTURA', 'CONTRATO', 'CONCEPTO','TIPO OP', 'FDO OP', 'FECHA', 'IMPORTE', 'ETIQUETA','MES']]
```

## Analisis multidimensional

```python
compras_df.dtypes
```

```python
# Agrupar los datos por el beneficiario y sumar los importes
sum_importe_beneficiario = compras_df.groupby('BENEFICIARIO')['IMPORTE'].sum().reset_index()
```

```python
sum_importe_beneficiario.head(5)
```

```python
sum_importe_beneficiario.reset_index().sort_values(by='IMPORTE', ascending=False).iloc[0:10].plot(kind='bar', x='BENEFICIARIO',y='IMPORTE')
```

```python
mean_importe_sheet=compras_df.groupby('ETIQUETA')['IMPORTE'].mean()
```

```python
mean_importe_sheet.head(4)
```

```python
mean_importe_sheet.reset_index().sort_values(by='IMPORTE', ascending=False).iloc[0:10].plot(kind='bar', x='ETIQUETA',y='IMPORTE')
```

```python
sum_importe_mes=compras_df.groupby('MES')['IMPORTE'].sum()
```

```python
sum_importe_mes.reset_index().sort_values(by='MES', ascending=True).plot(kind='bar', x='MES',y='IMPORTE')
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
texto_compras=''
```

```python
for row in compras_df['CONCEPTO']:
    texto_compras=texto_compras+ ' '+ row
```

```python
stopwords_list= stopwords.words('spanish')+['2023', 'cont', 'millar', 'dpp', 'diciembre','municipio', 'león']
```

```python
texto_compras = ' '.join([x for x in texto_compras.strip().lower().split(' ') if x not in stopwords_list and len(x)>4])
```

```python
wc = WordCloud(background_color= 'white')
wc.generate(texto_compras)
```

```python
plt.axis("off")
plt.imshow(wc, interpolation='bilinear')
plt.show()
```
